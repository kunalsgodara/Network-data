```python
import sys
import os

sys.path.append(os.path.abspath(".."))

import networkx as nx
import random
import matplotlib.pyplot as plt

from src.simulations import simulate_spread
```

```python
small_graph = nx.read_gml("../outputs/graphs/small_graph.gml")
medium_graph = nx.read_gml("../outputs/graphs/medium_graph.gml")
large_graph = nx.read_gml("../outputs/graphs/large_graph.gml")
```

```python
def get_lcc(G):
    largest_cc = max(nx.connected_components(G), key=len)
    return G.subgraph(largest_cc).copy()

small_lcc = get_lcc(small_graph)
medium_lcc = get_lcc(medium_graph)
large_lcc = get_lcc(large_graph)
```

```python
random_start = random.choice(list(small_lcc.nodes()))

spread_random = simulate_spread(small_lcc, random_start)

print("Random start node:", random_start)
print("Spread history:", spread_random)
```

```python
plt.plot(spread_random)
plt.xlabel("Steps")
plt.ylabel("Total Infected")
plt.title("Spread from Random User")
plt.show()
```

```python
highest_degree_node = max(small_lcc.degree, key=lambda x: x[1])[0]

spread_degree = simulate_spread(small_lcc, highest_degree_node)

print("Highest degree node:", highest_degree_node)
print("Spread history:", spread_degree)
```

```python
plt.plot(spread_degree)
plt.xlabel("Steps")
plt.ylabel("Total Infected")
plt.title("Spread from Highest Degree User")
plt.show()
```

```python
plt.plot(spread_random, label="Random Start")
plt.plot(spread_degree, label="High Degree Start")

plt.xlabel("Steps")
plt.ylabel("Total Infected")
plt.title("Spread Comparison")
plt.legend()

plt.show()
```

```python
betweenness = nx.betweenness_centrality(small_lcc)

highest_bc_node = max(betweenness, key=betweenness.get)

print("Highest Betweenness Node:", highest_bc_node)
```

```python
spread_bc = simulate_spread(small_lcc, highest_bc_node)

print("Spread history:", spread_bc)
```

```python
plt.plot(spread_random, label="Random Start")
plt.plot(spread_degree, label="High Degree Start")
plt.plot(spread_bc, label="High Betweenness Start")

plt.xlabel("Steps")
plt.ylabel("Total Infected")
plt.title("Spread Strategy Comparison")

plt.legend()
plt.show()
```

```python
import numpy as np

def average_spread(G, start_node, runs=50, beta=0.2, steps=20):

    all_runs = []

    for _ in range(runs):
        spread = simulate_spread(G, start_node, beta=beta, steps=steps)
        all_runs.append(spread)

    max_len = max(len(r) for r in all_runs)

    # pad shorter runs
    padded = []
    for r in all_runs:
        padded.append(r + [r[-1]] * (max_len - len(r)))

    return np.mean(padded, axis=0)
```

```python
avg_random = average_spread(large_lcc, random_start)
avg_degree = average_spread(large_lcc, highest_degree_node)
avg_bc = average_spread(large_lcc, highest_bc_node)
```

```python
plt.plot(avg_random, label="Random Start")
plt.plot(avg_degree, label="High Degree Start")
plt.plot(avg_bc, label="High Betweenness Start")

plt.xlabel("Steps")
plt.ylabel("Average Infected")
plt.title("Average Spread Over 50 Simulations")

plt.legend()
plt.show()
```
